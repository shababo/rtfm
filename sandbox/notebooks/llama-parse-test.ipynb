```python
import requests
import os
from glob import glob
from pprint import pprint

from dotenv import load_dotenv
load_dotenv()

import nest_asyncio
nest_asyncio.apply()

from llama_parse import LlamaParse
from llama_index.core import SimpleDirectoryReader
```

```python
job_id = "81d481dc-a722-43b6-aa73-c88e399aec0e"
```

```python
url = f"https://api.cloud.llamaindex.ai/api/v1/parsing/job/{job_id}/details"
```

```python
payload={}
headers = {
  'Accept': 'application/json',
  'Authorization': f'Bearer {os.environ.get("LLAMA_CLOUD_API_KEY")}',
  'Authorization': f'Bearer {os.environ.get("LLAMA_CLOUD_API_KEY")}'
}

response = requests.request("GET", url, headers=headers, data=payload)
```

```python
# set up parser
parser = LlamaParse(
    result_type="markdown"  # "markdown" and "text" are available
)
```

```python
# use SimpleDirectoryReader to parse our file
file_extractor = {".pdf": parser}
documents = SimpleDirectoryReader(input_dir='../../manuals/dmd', file_extractor=file_extractor).load_data()
print(documents)
```
